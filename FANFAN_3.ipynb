```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import UpSampling1D, Layer, Permute, SpatialDropout1D, TimeDistributed, Embedding, Concatenate, Activation, Multiply, Bidirectional, LSTM, Add, Conv1D, Dense, Flatten, Dropout, BatchNormalization, LayerNormalization, MaxPooling1D, GlobalAveragePooling1D, GlobalMaxPooling1D, Lambda, Reshape, Multiply, Input, Softmax, MultiHeadAttention
from tensorflow.keras.optimizers import Adam, Nadam, Adafactor
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, CSVLogger, Callback
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from tensorflow.keras.regularizers import l2, l1_l2, l1
from sklearn.preprocessing import RobustScaler
from tensorflow.keras.layers import MultiHeadAttention
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.losses import BinaryFocalCrossentropy
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import constraints, activations, initializers, regularizers
from tensorflow.keras.constraints import NonNeg
from tensorflow.keras.constraints import Constraint

# Check available GPUs
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

# Define constants
EPOCHS = 999
BATCH_SIZE = 16 # 32 Test Accuracy: 0.8842 with class weights without custom layers
INITIAL_LEARNING_RATE = 0.0005  #*(BATCH_SIZE/32)
MODEL_FILE_PATH = "best_model_file_FANFAN_3.keras"

class RevertToBestWithPatience(tf.keras.callbacks.Callback):
    def __init__(self, patience=2):
        super(RevertToBestWithPatience, self).__init__()
        self.patience = patience
        self.wait = 0
        self.best_val_loss = np.Inf
        self.best_weights = None

    def on_train_begin(self, logs=None):
        self.wait = 0
        self.best_val_loss = np.Inf
        self.best_weights = None

    def on_epoch_end(self, epoch, logs=None):
        val_loss = logs.get('val_loss')
        if val_loss < self.best_val_loss:
            # Improvement found
            self.best_val_loss = val_loss
            self.best_weights = self.model.get_weights()
            self.wait = 0  # reset wait counter
        else:
            # No improvement this epoch
            self.wait += 1
            if self.wait >= self.patience and self.best_weights is not None:
                # Patience exceeded, revert to best weights
                self.model.set_weights(self.best_weights)
                print(f"Reverted to best weights at epoch {epoch+1}, with val_loss = {self.best_val_loss}")
                self.wait = 0  # reset counter after reverting

# Define callbacks
checkpoint = ModelCheckpoint(MODEL_FILE_PATH, monitor='val_loss', save_best_only=True, mode='min')
earlystop = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.0000625)
csv_logger = CSVLogger('training_FANFAN_3.csv')

# Create a list of callbacks
callbacks = [earlystop, csv_logger, reduce_lr, checkpoint, RevertToBestWithPatience(patience=7)]

class FANLayer(tf.keras.layers.Layer):
    """
    FANLayer: The layer used in FAN (https://arxiv.org/abs/2410.02675).

    Args:
        input_dim (int): The number of input features.
        output_dim (int): The number of output features.
        p_ratio (float): The ratio of output dimensions used for cosine and sine parts (default: 0.25).
        activation (str or callable): The activation function to apply to the g component (default: 'gelu').
        use_p_bias (bool): If True, include bias in the linear transformations of the p component (default: True).
        gated (bool): If True, applies gating to the output.
        kernel_regularizer: Regularizer for kernel weights.
        bias_regularizer: Regularizer for bias weights.
    """

    def __init__(self,
                 output_dim,
                 p_ratio=0.25,
                 activation='gelu',
                 use_p_bias=True,
                 gated=False,
                 kernel_regularizer=None,
                 bias_regularizer=None,
                 **kwargs):
        super(FANLayer, self).__init__(**kwargs)

        assert 0 < p_ratio < 0.5, "p_ratio must be between 0 and 0.5"

        self.p_ratio = p_ratio
        self.output_dim = output_dim
        self.activation = activations.get(activation)
        self.use_p_bias = use_p_bias
        self.gated = gated
        self.kernel_regularizer = regularizers.get(kernel_regularizer)
        self.bias_regularizer = regularizers.get(bias_regularizer)

        # Compute output dimensions for p and g components
        self.p_output_dim = int(output_dim * self.p_ratio)
        self.g_output_dim = output_dim - 2 * self.p_output_dim  # Account for cosine and sine

        # Layers for linear transformations
        self.input_linear_p = layers.Dense(self.p_output_dim,
                                    use_bias=self.use_p_bias,
                                    kernel_regularizer=self.kernel_regularizer,
                                    bias_regularizer=self.bias_regularizer)
        self.input_linear_g = layers.Dense(self.g_output_dim,
                                    kernel_regularizer=self.kernel_regularizer,
                                    bias_regularizer=self.bias_regularizer)

        if self.gated:
            self.gate = self.add_weight(name='gate',
                                        shape=(1,),
                                        initializer=initializers.RandomNormal(),
                                        trainable=True,
                                            regularizer=None,
                                            constraint=NonNeg())

    def call(self, inputs):
        # Apply the linear transformation followed by the activation for the g component
        g = self.activation(self.input_linear_g(inputs))

        # Apply the linear transformation for the p component
        p = self.input_linear_p(inputs)

        if self.gated:
            gate = tf.sigmoid(self.gate)
            output = tf.concat([gate * tf.cos(p), gate * tf.sin(p), (1 - gate) * g], axis=-1)
        else:
            output = tf.concat([tf.cos(p), tf.sin(p), g], axis=-1)

        return output

    def get_config(self):
        config = super(FANLayer, self).get_config()
        config.update({
            "output_dim": self.output_dim,
            "p_ratio": self.p_ratio,
            "activation": activations.serialize(self.activation),
            "use_p_bias": self.use_p_bias,
            "gated": self.gated,
            "kernel_regularizer": regularizers.serialize(self.kernel_regularizer),
            "bias_regularizer": regularizers.serialize(self.bias_regularizer)
        })
        return config

    def compute_output_shape(self, input_shape):
        """Computes the output shape of the layer.

        Args:
            input_shape (tuple): Shape tuple (tuple of integers)
                of the input tensor.

        Returns:
            tuple: Output shape tuple.
        """
        return (input_shape[0], self.output_dim)

def preprocess_data(data, scaler=True):
    grouped = data.groupby('Name')
    points_list = []
    category_list = []
    labels_list = []

    for name, group in grouped:
        points = group[group.columns[4:]].values.astype(np.float32)

        if points.shape != (6, 588):
            print(f"Skipping group {name} due to unexpected shape: {points.shape}")
            continue

        points = np.expand_dims(points, axis=-1)  # Reshape to (6, 588, 1)

        if scaler==True:
          # Normalize each sample independently
          scaler = RobustScaler()
          points = points.reshape(-1, 1)
          points = scaler.fit_transform(points)
          points = points.reshape(6, 588, 1)

        points_list.append(points)
        category_list.append(group['Time'].values.astype(np.float32))
        labels_list.append(group['Covid.Nominal'].values.astype(np.float32))

        points = np.array(points_list)
        category = np.array(category_list) - 1  # Adjust if 'Time' starts from 1
        labels = np.array(labels_list)

        # Check if labels are consistent within each patient
        inconsistent_labels = np.any(labels != labels[:, [0]], axis=1)
        if np.any(inconsistent_labels):
          print("Warning: Inconsistent labels found for some patients.")
          # You can decide how to handle this, e.g., skip these patients
          consistent_indices = ~inconsistent_labels
          points = points[consistent_indices]
          category = category[consistent_indices]
          labels = labels[consistent_indices]

        category = np.expand_dims(category, axis=-1)  # Reshape to (6, 1)
        labels = np.expand_dims(labels, axis=-1)  # Reshape to (6, 1)
        # Aggregate labels to a single label per patient
        #labels = labels[:, 0]  # Take the first label
        #labels = labels.reshape(-1, 1)  # Shape: (num_samples, 1)

    return points, category, labels



# Load and preprocess data
data = pd.read_csv('20210329_train.csv')
points, category, labels = preprocess_data(data, scaler=False)
data_test=pd.read_csv('20210329_test.csv')
points_test, category_test, labels_test = preprocess_data(data_test, scaler=False)

print("Original data shapes:")
print(f"Points: {points.shape}")
print(f"Category: {category.shape}")
print(f"Labels: {labels.shape}")
print("Test data shapes:")
print(f"Points: {points_test.shape}")
print(f"Category: {category_test.shape}")
print(f"Labels: {labels_test.shape}")

# Split data into training and validation sets
X_train, X_val, z_train, z_val, y_train, y_val = train_test_split(points, category, labels, test_size=0.2, random_state=42)

# Convert to tf.data.Dataset
train = tf.data.Dataset.from_tensor_slices(((X_train,z_train), y_train))
train = train.batch(BATCH_SIZE)
train = train.cache()
train = train.shuffle(buffer_size=train.cardinality(), reshuffle_each_iteration=True).prefetch(tf.data.AUTOTUNE)

test = tf.data.Dataset.from_tensor_slices(((X_val, z_val), y_val))
test = test.batch(BATCH_SIZE)
test = test.cache()
test = test.shuffle(buffer_size=test.cardinality(), reshuffle_each_iteration=True).prefetch(tf.data.AUTOTUNE)

testtt = tf.data.Dataset.from_tensor_slices(((points_test, category_test), labels_test))
testtt = testtt.batch(BATCH_SIZE)
testtt = testtt.cache()
testtt = testtt.shuffle(buffer_size=testtt.cardinality(), reshuffle_each_iteration=True).prefetch(tf.data.AUTOTUNE)

# Extract labels to compute class weights
labels_list = []

# Extract all labels from the dataset (in batches, if batched)
for _, labels in train:
    labels_list.extend(labels.numpy().reshape(-1))

# Convert the list of labels to a numpy array
labels_array = np.array(labels_list)

# Calculate class weights using sklearn's compute_class_weight
class_weights_dict = compute_class_weight('balanced', classes=np.unique(labels_array), y=labels_array)

# Convert class weights to a dictionary
class_weights = {i: class_weights_dict[i] for i in range(len(class_weights_dict))}

print(f"Class Weights: {class_weights}")


# Define the input shapes
time_steps = 6
feature_dim = 588
point_input_shape = (time_steps, feature_dim, 1)
category_input_shape = (time_steps, 1)

# Update input shapes for TimeDistributed layers
#point_input_shape = (6, 588, 1)  # 6 measurements per patient
#category_input_shape = (6, 1)    # Time categories for each measurement

def create_model(point_input_shape, category_input_shape):
    # Inputs
    point_input = Input(shape=point_input_shape, name='point_input')
    category_input = Input(shape=category_input_shape, name='category_input')

    # Convolutional Layers with TimeDistributed Conv1D

    # First Conv Block
    x = TimeDistributed(Conv1D(128, 5, activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(point_input)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(MaxPooling1D(2))(x)
    x = TimeDistributed(SpatialDropout1D(0.3))(x)

    # Save residual from the first block
    residual_1 = x

    # Second Conv Block
    x = TimeDistributed(Conv1D(256, 5, activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(x)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(MaxPooling1D(2))(x)
    x = TimeDistributed(SpatialDropout1D(0.3))(x)

    # Match the pooling for residual_1
    residual_1 = TimeDistributed(MaxPooling1D(2))(residual_1)

    # Align residual dimensions using 1x1 convolution to match the new feature size (256)
    residual_1 = TimeDistributed(Conv1D(256, 1, padding='same'))(residual_1)

    # Add residual connection from the first block to the second block
    x = Add()([x, residual_1])

    # Save residual from the second block
    residual_2 = x

    # Third Conv Block
    x = TimeDistributed(Conv1D(512, 5, activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(x)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(MaxPooling1D(2))(x)


    # Match the pooling for residual_2
    residual_2 = TimeDistributed(MaxPooling1D(2))(residual_2)

    # Align residual dimensions using 1x1 convolution to match the new feature size (512)
    residual_2 = TimeDistributed(Conv1D(512, 1, padding='same'))(residual_2)

    # Add residual connection from the second block to the third block
    x = Add()([x, residual_2])

    # Channel Attention using TimeDistributed pooling layers
    avg_pool = TimeDistributed(GlobalAveragePooling1D())(x)  # Shape: (batch_size, time_steps, channels)
    max_pool = TimeDistributed(GlobalMaxPooling1D())(x)  # Shape: (batch_size, time_steps, channels)

    # Shared Dense Layers
    shared_layer_one = TimeDistributed(Dense(256, activation='relu', kernel_initializer='he_normal', use_bias=True))
    shared_layer_two = TimeDistributed(Dense(512, kernel_initializer='he_normal', use_bias=True))

    # Channel Attention Block
    avg_pool = shared_layer_one(avg_pool)
    max_pool = shared_layer_one(max_pool)
    cbam_feature = Concatenate()([avg_pool, max_pool])
    cbam_feature = shared_layer_two(cbam_feature)
    cbam_feature = Activation('sigmoid')(cbam_feature)

    # Reshape cbam_feature to match x's shape using Lambda layer
    conv1d_output_shape = x.shape
    cbam_feature = TimeDistributed(Lambda(lambda t: tf.expand_dims(t, axis=1), output_shape=lambda s: (s[0], s[1], 1)))(cbam_feature)
    cbam_feature = TimeDistributed(Lambda(lambda t: tf.repeat(t, repeats=conv1d_output_shape[2], axis=1), output_shape=lambda s: (s[0], conv1d_output_shape[2], s[1])))(cbam_feature)

    # Residual Connection after Channel Attention
    print(f"residual_1 shape: {residual_1.shape}")
    print(f"x shape: {x.shape}")
    print(f"cbam_feature shape: {cbam_feature.shape}")

    # Align time dimension between x and residual_1 using upsampling/pooling until matching
    while True:
        residual_time_dim = residual_1.shape[2]
        x_time_dim = x.shape[2]

        if residual_time_dim == x_time_dim:
            break
        elif residual_time_dim > x_time_dim:
            # Reduce residual time dimension by pooling
            residual_1 = TimeDistributed(MaxPooling1D(pool_size=2))(residual_1)
        elif residual_time_dim < x_time_dim:
            # Increase residual time dimension by upsampling
            residual_1 = TimeDistributed(UpSampling1D(size=2))(residual_1)

    # Residual Connection after Channel Attention
    print(f"residual_1 shape: {residual_1.shape}")
    print(f"x shape: {x.shape}")
    print(f"cbam_feature shape: {cbam_feature.shape}")

    # Align feature dimensions using Conv1D if necessary
    if residual_1.shape[-1] != x.shape[-1]:
        residual_1 = TimeDistributed(Conv1D(x.shape[-1], 1, padding='same'))(residual_1)

    # Residual Connection after Channel Attention
    print(f"residual_1 shape: {residual_1.shape}")
    print(f"x shape: {x.shape}")
    print(f"cbam_feature shape: {cbam_feature.shape}")

    x = Multiply()([x, cbam_feature])
    x = Add()([x, residual_1])  # Add residual from first Conv block after matching dimensions


    # Spatial Attention
    avg_pool = TimeDistributed(Lambda(lambda x: tf.reduce_mean(x, axis=-1, keepdims=True), output_shape=lambda s: (s[0], s[1], 1)))(x)
    max_pool = TimeDistributed(Lambda(lambda x: tf.reduce_max(x, axis=-1, keepdims=True), output_shape=lambda s: (s[0], s[1], 1)))(x)

    concat = Concatenate(axis=-1)([avg_pool, max_pool])
    spatial_attention_feature = TimeDistributed(Conv1D(1, 7, padding='same', activation='sigmoid', kernel_initializer='he_normal', use_bias=False))(concat)
    x = Multiply()([x, spatial_attention_feature])

    # Flatten the attention weights for LSTM layers
    flattened_attention = Reshape((x.shape[1], -1))(x)

    # Reduce dimensionality of flattened_attention
    reduced_attention = Dense(512, activation='relu')(flattened_attention)

    # LSTM for Temporal Modeling using TimeDistributed
    x = Bidirectional(LSTM(256, return_sequences=True, dropout=0.33, recurrent_dropout=0.33))(flattened_attention)
    x = Add()([x, reduced_attention])  # Residual connection

    # Another LSTM layer with residual connection
    x_residual = x
    x = Bidirectional(LSTM(256, return_sequences=True, dropout=0.33, recurrent_dropout=0.33))(x)
    x = Add()([x, x_residual])
    x = Dropout(0.3)(x)

    # Temporal Attention Mechanism using Multi-Head Self-Attention
    attention_output = MultiHeadAttention(num_heads=6, key_dim=128)(x, x)
    attention_weights = TimeDistributed(Activation('sigmoid'))(attention_output)
    x = Multiply()([x, attention_weights])

    # Residual Connection after Attention Mechanism
    x = Add()([x, x_residual])

    x = TimeDistributed(FANLayer(output_dim=512, p_ratio=0.25, activation='gelu', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(x)

    #x = Add()([x, fan_res])

    # Handle category_input with Embedding
    embedding_dim = 64
    num_categories = 6  # Replace with the actual number of categories

    embedding_layer = TimeDistributed(Embedding(input_dim=num_categories, output_dim=embedding_dim, name='embedding'))(category_input)
    category_embedded = Reshape((-1, embedding_layer.shape[-1] * embedding_layer.shape[-2]))(embedding_layer)

    # Apply Dense layer to adjust dimensionality (if needed)
    category_embedded = TimeDistributed(Dense(x.shape[-1], activation='relu'))(category_embedded)

    # Concatenate Point and Category Representations
    concatenated = Concatenate(axis=-1)([x, category_embedded])

    # Apply dense layers with regularization
    dense_layer = Dense(128, activation='relu', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4))(concatenated)
    dense_layer = BatchNormalization()(dense_layer)
    dropout = Dropout(0.5)(dense_layer)

    outputs = TimeDistributed(Dense(1, activation='sigmoid'))(dropout)

    # Create Model
    model = Model(inputs=[point_input, category_input], outputs=outputs)
    return model

model = create_model(point_input_shape,category_input_shape)
optimizer = Adam(learning_rate=INITIAL_LEARNING_RATE, clipnorm=0.5)
loss = BinaryFocalCrossentropy(apply_class_balancing=False,
                              alpha=0.5869,
                              gamma=2.0,
                              from_logits=False,
                              label_smoothing=0.0,
                              axis=-1,
                              reduction='sum_over_batch_size',
                              name='binary_focal_crossentropy'
                              )
model.compile(optimizer=optimizer, loss=loss, metrics=["accuracy", tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])
print(model.summary())


# Train the model
history = model.fit(train,
                    epochs=EPOCHS,
                    validation_data=test, # Pass both for validation
                    verbose=1,
                    callbacks=callbacks,
                    class_weight=class_weights,
                    )
# Save the model
model.save("cbam_trial_FANFAN_3.keras")
model.save("cbam_trial_FANFAN_3.h5")


# Plot training history
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax = ax.ravel()
for i, metric in enumerate(['accuracy', 'loss']):
    ax[i].plot(history.history[metric])
    ax[i].plot(history.history['val_' + metric])
    ax[i].set_title(f'Model {metric}')
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(metric)
    ax[i].legend(['train', 'val'])
plt.savefig("training_history.png")


# Evaluate the model on the test data
evaluation = model.evaluate(testtt)
print(f"Test Loss: {evaluation[0]}, Test Accuracy: {evaluation[1]}")

# Pad the dataset to ensure all batches have the same size
padded_testtt = testtt.padded_batch(len(testtt))

# Convert the padded dataset to numpy arrays in a single step
(X_test, z_test), y_test = tf.data.experimental.get_single_element(padded_testtt)

# Convert tensors to numpy arrays
X_test = X_test.numpy()
z_test = z_test.numpy()
y_test = y_test.numpy()

# Print shapes for debugging purposes
print(f"Shape of X_test after padding: {X_test.shape}")
print(f"Shape of z_test after padding: {z_test.shape}")
print(f"Shape of y_test after padding: {y_test.shape}")

# Reshape to remove padding dimensions
# Original shape after padding: (22, 16, 6, 588, 1)
# Flatten first two dimensions (e.g., 22 * 16)
X_test = X_test.reshape(-1, 6, 588, 1)  # Shape: (22*16, 6, 588, 1) = (352, 6, 588, 1)
z_test = z_test.reshape(-1, 6, 1)       # Shape: (22*16, 6, 1) = (352, 6, 1)
y_test = y_test.reshape(-1, 6, 1)       # Shape: (22*16, 6, 1) = (352, 6, 1)

# Print reshaped shapes for debugging purposes
print(f"Reshaped X_test: {X_test.shape}")
print(f"Reshaped z_test: {z_test.shape}")
print(f"Reshaped y_test: {y_test.shape}")

# Make predictions using the model
y_pred = model.predict([X_test, z_test])  # Pass both points and category inputs
print(f"Shape of y_pred: {y_pred.shape}")

# Calculate mean values across the time step dimension
# Assuming your y_pred has shape (number_of_samples, 6, 1)
mean_values = np.mean(np.squeeze(y_pred), axis=1)  # Calculate the average across the time steps
print(f"Shape of mean_values: {mean_values.shape}")

# Threshold Optimization to find the best threshold
thresholds = np.arange(0.01, 0.99, 0.001)
best_f1 = 0
best_threshold = 0.5

for threshold in thresholds:
    y_pred_binary = (mean_values > threshold).astype(int)
    f1 = f1_score(y_test[:, 0], y_pred_binary, average='weighted')
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold

print(f"Best Threshold: {best_threshold}, Best F1 Score: {best_f1:.4f}")

# Convert mean_values to binary predictions using the best threshold
y_pred_binary = (mean_values > best_threshold).astype(int)

# Ensure y_test contains binary labels (if necessary)
y_test_binary = y_test[:, 0]

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_binary, y_pred_binary)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.savefig("confusion_matrix.png")

# F1 Score and Classification Report
f1 = f1_score(y_test_binary, y_pred_binary, average='weighted')
print(f"F1 Score: {f1:.4f}")
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_binary))

# Save classification report and metrics to a file
with open("evaluation_report_3.txt", "w") as f:
    f.write(f"Best Threshold: {best_threshold}, Best F1 Score: {best_f1:.4f}\n")
    f.write(f"F1 Score with Best Threshold: {f1:.4f}\n")
    f.write("Classification Report:\n")
    f.write(classification_report(y_test_binary, y_pred_binary))

# Plot Precision-Recall Curve (Optional)
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test_binary, mean_values)
plt.figure(figsize=(10, 7))
plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Precision/Recall')
plt.title('Precision-Recall vs Threshold')
plt.legend()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import UpSampling1D, Layer, Permute, SpatialDropout1D, TimeDistributed, Embedding, Concatenate, Activation, Multiply, Bidirectional, LSTM, Add, Conv1D, Dense, Flatten, Dropout, BatchNormalization, LayerNormalization, MaxPooling1D, GlobalAveragePooling1D, GlobalMaxPooling1D, Lambda, Reshape, Multiply, Input, Softmax, MultiHeadAttention
from tensorflow.keras.optimizers import Adam, Nadam, Adafactor
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, CSVLogger, Callback
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from tensorflow.keras.regularizers import l2, l1_l2, l1
from sklearn.preprocessing import RobustScaler
from tensorflow.keras.layers import MultiHeadAttention
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.losses import BinaryFocalCrossentropy
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import constraints, activations, initializers, regularizers
from tensorflow.keras.constraints import NonNeg
from tensorflow.keras.constraints import Constraint

# Check available GPUs
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

# Define constants
EPOCHS = 999
BATCH_SIZE = 32 # 32 Test Accuracy: 0.8842 with class weights without custom layers
INITIAL_LEARNING_RATE = 0.0005  #*(BATCH_SIZE/32)
MODEL_FILE_PATH = "best_model_file_FANFAN_3.keras"

class RevertToBestWithPatience(tf.keras.callbacks.Callback):
    def __init__(self, patience=2):
        super(RevertToBestWithPatience, self).__init__()
        self.patience = patience
        self.wait = 0
        self.best_val_loss = np.Inf
        self.best_weights = None

    def on_train_begin(self, logs=None):
        self.wait = 0
        self.best_val_loss = np.Inf
        self.best_weights = None

    def on_epoch_end(self, epoch, logs=None):
        val_loss = logs.get('val_loss')
        if val_loss < self.best_val_loss:
            # Improvement found
            self.best_val_loss = val_loss
            self.best_weights = self.model.get_weights()
            self.wait = 0  # reset wait counter
        else:
            # No improvement this epoch
            self.wait += 1
            if self.wait >= self.patience and self.best_weights is not None:
                # Patience exceeded, revert to best weights
                self.model.set_weights(self.best_weights)
                print(f"Reverted to best weights at epoch {epoch+1}, with val_loss = {self.best_val_loss}")
                self.wait = 0  # reset counter after reverting

# Define callbacks
checkpoint = ModelCheckpoint(MODEL_FILE_PATH, monitor='val_loss', save_best_only=True, mode='min')
earlystop = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00000125)
csv_logger = CSVLogger('training_FANFAN_3.csv')

# Create a list of callbacks
callbacks = [earlystop, csv_logger, reduce_lr, checkpoint, RevertToBestWithPatience(patience=7)]

class FANLayer(tf.keras.layers.Layer):
    """
    FANLayer: The layer used in FAN (https://arxiv.org/abs/2410.02675).

    Args:
        input_dim (int): The number of input features.
        output_dim (int): The number of output features.
        p_ratio (float): The ratio of output dimensions used for cosine and sine parts (default: 0.25).
        activation (str or callable): The activation function to apply to the g component (default: 'gelu').
        use_p_bias (bool): If True, include bias in the linear transformations of the p component (default: True).
        gated (bool): If True, applies gating to the output.
        kernel_regularizer: Regularizer for kernel weights.
        bias_regularizer: Regularizer for bias weights.
    """

    def __init__(self,
                 output_dim,
                 p_ratio=0.25,
                 activation='gelu',
                 use_p_bias=True,
                 gated=False,
                 kernel_regularizer=None,
                 bias_regularizer=None,
                 **kwargs):
        super(FANLayer, self).__init__(**kwargs)

        assert 0 < p_ratio < 0.5, "p_ratio must be between 0 and 0.5"

        self.p_ratio = p_ratio
        self.output_dim = output_dim
        self.activation = activations.get(activation)
        self.use_p_bias = use_p_bias
        self.gated = gated
        self.kernel_regularizer = regularizers.get(kernel_regularizer)
        self.bias_regularizer = regularizers.get(bias_regularizer)

        # Compute output dimensions for p and g components
        self.p_output_dim = int(output_dim * self.p_ratio)
        self.g_output_dim = output_dim - 2 * self.p_output_dim  # Account for cosine and sine

        # Layers for linear transformations
        self.input_linear_p = layers.Dense(self.p_output_dim,
                                    use_bias=self.use_p_bias,
                                    kernel_regularizer=self.kernel_regularizer,
                                    bias_regularizer=self.bias_regularizer)
        self.input_linear_g = layers.Dense(self.g_output_dim,
                                    kernel_regularizer=self.kernel_regularizer,
                                    bias_regularizer=self.bias_regularizer)

        if self.gated:
            self.gate = self.add_weight(name='gate',
                                        shape=(1,),
                                        initializer=initializers.RandomNormal(),
                                        trainable=True,
                                            regularizer=None,
                                            constraint=NonNeg())

    def call(self, inputs):
        # Apply the linear transformation followed by the activation for the g component
        g = self.activation(self.input_linear_g(inputs))

        # Apply the linear transformation for the p component
        p = self.input_linear_p(inputs)

        if self.gated:
            gate = tf.sigmoid(self.gate)
            output = tf.concat([gate * tf.cos(p), gate * tf.sin(p), (1 - gate) * g], axis=-1)
        else:
            output = tf.concat([tf.cos(p), tf.sin(p), g], axis=-1)

        return output

    def get_config(self):
        config = super(FANLayer, self).get_config()
        config.update({
            "output_dim": self.output_dim,
            "p_ratio": self.p_ratio,
            "activation": activations.serialize(self.activation),
            "use_p_bias": self.use_p_bias,
            "gated": self.gated,
            "kernel_regularizer": regularizers.serialize(self.kernel_regularizer),
            "bias_regularizer": regularizers.serialize(self.bias_regularizer)
        })
        return config

    def compute_output_shape(self, input_shape):
        """Computes the output shape of the layer.

        Args:
            input_shape (tuple): Shape tuple (tuple of integers)
                of the input tensor.

        Returns:
            tuple: Output shape tuple.
        """
        return (input_shape[0], self.output_dim)

def preprocess_data(data, scaler=True):
    grouped = data.groupby('Name')
    points_list = []
    category_list = []
    labels_list = []

    for name, group in grouped:
        points = group[group.columns[4:]].values.astype(np.float32)

        if points.shape != (6, 588):
            print(f"Skipping group {name} due to unexpected shape: {points.shape}")
            continue

        points = np.expand_dims(points, axis=-1)  # Reshape to (6, 588, 1)

        if scaler==True:
          # Normalize each sample independently
          scaler = RobustScaler()
          points = points.reshape(-1, 1)
          points = scaler.fit_transform(points)
          points = points.reshape(6, 588, 1)

        points_list.append(points)
        category_list.append(group['Time'].values.astype(np.float32))
        labels_list.append(group['Covid.Nominal'].values.astype(np.float32))

        points = np.array(points_list)
        category = np.array(category_list) - 1  # Adjust if 'Time' starts from 1
        labels = np.array(labels_list)

        # Check if labels are consistent within each patient
        inconsistent_labels = np.any(labels != labels[:, [0]], axis=1)
        if np.any(inconsistent_labels):
          print("Warning: Inconsistent labels found for some patients.")
          # You can decide how to handle this, e.g., skip these patients
          consistent_indices = ~inconsistent_labels
          points = points[consistent_indices]
          category = category[consistent_indices]
          labels = labels[consistent_indices]

        category = np.expand_dims(category, axis=-1)  # Reshape to (6, 1)
        labels = np.expand_dims(labels, axis=-1)  # Reshape to (6, 1)
        # Aggregate labels to a single label per patient
        #labels = labels[:, 0]  # Take the first label
        #labels = labels.reshape(-1, 1)  # Shape: (num_samples, 1)

    return points, category, labels



# Load and preprocess data
data = pd.read_csv('20210329_train.csv')
points, category, labels = preprocess_data(data, scaler=False)
data_test=pd.read_csv('20210329_test.csv')
points_test, category_test, labels_test = preprocess_data(data_test, scaler=False)

print("Original data shapes:")
print(f"Points: {points.shape}")
print(f"Category: {category.shape}")
print(f"Labels: {labels.shape}")
print("Test data shapes:")
print(f"Points: {points_test.shape}")
print(f"Category: {category_test.shape}")
print(f"Labels: {labels_test.shape}")

# Split data into training and validation sets
X_train, X_val, z_train, z_val, y_train, y_val = train_test_split(points, category, labels, test_size=0.2, random_state=42)

# Convert to tf.data.Dataset
train = tf.data.Dataset.from_tensor_slices(((X_train,z_train), y_train))
train = train.batch(BATCH_SIZE)
train = train.cache()
train = train.shuffle(buffer_size=train.cardinality(), reshuffle_each_iteration=True).prefetch(tf.data.AUTOTUNE)

test = tf.data.Dataset.from_tensor_slices(((X_val, z_val), y_val))
test = test.batch(BATCH_SIZE)
test = test.cache()
test = test.shuffle(buffer_size=test.cardinality(), reshuffle_each_iteration=True).prefetch(tf.data.AUTOTUNE)

testtt = tf.data.Dataset.from_tensor_slices(((points_test, category_test), labels_test))
testtt = testtt.batch(BATCH_SIZE)
testtt = testtt.cache()
testtt = testtt.shuffle(buffer_size=testtt.cardinality(), reshuffle_each_iteration=True).prefetch(tf.data.AUTOTUNE)

# Extract labels to compute class weights
labels_list = []

# Extract all labels from the dataset (in batches, if batched)
for _, labels in train:
    labels_list.extend(labels.numpy().reshape(-1))

# Convert the list of labels to a numpy array
labels_array = np.array(labels_list)

# Calculate class weights using sklearn's compute_class_weight
class_weights_dict = compute_class_weight('balanced', classes=np.unique(labels_array), y=labels_array)

# Convert class weights to a dictionary
class_weights = {i: class_weights_dict[i] for i in range(len(class_weights_dict))}

print(f"Class Weights: {class_weights}")


# Define the input shapes
time_steps = 6
feature_dim = 588
point_input_shape = (time_steps, feature_dim, 1)
category_input_shape = (time_steps, 1)

# Update input shapes for TimeDistributed layers
#point_input_shape = (6, 588, 1)  # 6 measurements per patient
#category_input_shape = (6, 1)    # Time categories for each measurement

def create_model(point_input_shape, category_input_shape):
    # Inputs
    point_input = Input(shape=point_input_shape, name='point_input')
    category_input = Input(shape=category_input_shape, name='category_input')

    # Convolutional Layers with TimeDistributed Conv1D

    # First Conv Block
    x = TimeDistributed(Conv1D(128, 5, activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(point_input)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(MaxPooling1D(2))(x)
    x = TimeDistributed(SpatialDropout1D(0.3))(x)

    # Save residual from the first block
    residual_1 = x

    # Second Conv Block
    x = TimeDistributed(Conv1D(256, 5, activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(x)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(MaxPooling1D(2))(x)
    x = TimeDistributed(SpatialDropout1D(0.3))(x)

    # Match the pooling for residual_1
    residual_1 = TimeDistributed(MaxPooling1D(2))(residual_1)

    # Align residual dimensions using 1x1 convolution to match the new feature size (256)
    residual_1 = TimeDistributed(Conv1D(256, 1, padding='same'))(residual_1)

    # Add residual connection from the first block to the second block
    x = Add()([x, residual_1])

    # Save residual from the second block
    residual_2 = x

    # Third Conv Block
    x = TimeDistributed(Conv1D(512, 5, activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(x)
    x = TimeDistributed(BatchNormalization())(x)
    x = TimeDistributed(MaxPooling1D(2))(x)


    # Match the pooling for residual_2
    residual_2 = TimeDistributed(MaxPooling1D(2))(residual_2)

    # Align residual dimensions using 1x1 convolution to match the new feature size (512)
    residual_2 = TimeDistributed(Conv1D(512, 1, padding='same'))(residual_2)

    # Add residual connection from the second block to the third block
    x = Add()([x, residual_2])

    # Channel Attention using TimeDistributed pooling layers
    avg_pool = TimeDistributed(GlobalAveragePooling1D())(x)  # Shape: (batch_size, time_steps, channels)
    max_pool = TimeDistributed(GlobalMaxPooling1D())(x)  # Shape: (batch_size, time_steps, channels)

    # Shared Dense Layers
    shared_layer_one = TimeDistributed(Dense(256, activation='relu', kernel_initializer='he_normal', use_bias=True))
    shared_layer_two = TimeDistributed(Dense(512, kernel_initializer='he_normal', use_bias=True))

    # Channel Attention Block
    avg_pool = shared_layer_one(avg_pool)
    max_pool = shared_layer_one(max_pool)
    cbam_feature = Concatenate()([avg_pool, max_pool])
    cbam_feature = shared_layer_two(cbam_feature)
    cbam_feature = Activation('sigmoid')(cbam_feature)

    # Reshape cbam_feature to match x's shape using Lambda layer
    conv1d_output_shape = x.shape
    cbam_feature = TimeDistributed(Lambda(lambda t: tf.expand_dims(t, axis=1), output_shape=lambda s: (s[0], s[1], 1)))(cbam_feature)
    cbam_feature = TimeDistributed(Lambda(lambda t: tf.repeat(t, repeats=conv1d_output_shape[2], axis=1), output_shape=lambda s: (s[0], conv1d_output_shape[2], s[1])))(cbam_feature)

    # Residual Connection after Channel Attention
    print(f"residual_1 shape: {residual_1.shape}")
    print(f"x shape: {x.shape}")
    print(f"cbam_feature shape: {cbam_feature.shape}")

    # Align time dimension between x and residual_1 using upsampling/pooling until matching
    while True:
        residual_time_dim = residual_1.shape[2]
        x_time_dim = x.shape[2]

        if residual_time_dim == x_time_dim:
            break
        elif residual_time_dim > x_time_dim:
            # Reduce residual time dimension by pooling
            residual_1 = TimeDistributed(MaxPooling1D(pool_size=2))(residual_1)
        elif residual_time_dim < x_time_dim:
            # Increase residual time dimension by upsampling
            residual_1 = TimeDistributed(UpSampling1D(size=2))(residual_1)

    # Residual Connection after Channel Attention
    print(f"residual_1 shape: {residual_1.shape}")
    print(f"x shape: {x.shape}")
    print(f"cbam_feature shape: {cbam_feature.shape}")

    # Align feature dimensions using Conv1D if necessary
    if residual_1.shape[-1] != x.shape[-1]:
        residual_1 = TimeDistributed(Conv1D(x.shape[-1], 1, padding='same'))(residual_1)

    # Residual Connection after Channel Attention
    print(f"residual_1 shape: {residual_1.shape}")
    print(f"x shape: {x.shape}")
    print(f"cbam_feature shape: {cbam_feature.shape}")

    x = Multiply()([x, cbam_feature])
    x = Add()([x, residual_1])  # Add residual from first Conv block after matching dimensions


    # Spatial Attention
    avg_pool = TimeDistributed(Lambda(lambda x: tf.reduce_mean(x, axis=-1, keepdims=True), output_shape=lambda s: (s[0], s[1], 1)))(x)
    max_pool = TimeDistributed(Lambda(lambda x: tf.reduce_max(x, axis=-1, keepdims=True), output_shape=lambda s: (s[0], s[1], 1)))(x)

    concat = Concatenate(axis=-1)([avg_pool, max_pool])
    spatial_attention_feature = TimeDistributed(Conv1D(1, 7, padding='same', activation='sigmoid', kernel_initializer='he_normal', use_bias=False))(concat)
    x = Multiply()([x, spatial_attention_feature])

    # Flatten the attention weights for LSTM layers
    flattened_attention = Reshape((x.shape[1], -1))(x)

    # Reduce dimensionality of flattened_attention
    reduced_attention = Dense(512, activation='relu')(flattened_attention)

    # LSTM for Temporal Modeling using TimeDistributed
    x = Bidirectional(LSTM(256, return_sequences=True, dropout=0.33, recurrent_dropout=0.33))(flattened_attention)
    x = Add()([x, reduced_attention])  # Residual connection

    # Another LSTM layer with residual connection
    x_residual = x
    x = Bidirectional(LSTM(256, return_sequences=True, dropout=0.33, recurrent_dropout=0.33))(x)
    x = Add()([x, x_residual])
    x = Dropout(0.3)(x)

    # Temporal Attention Mechanism using Multi-Head Self-Attention
    attention_output = MultiHeadAttention(num_heads=6, key_dim=128)(x, x)
    attention_weights = TimeDistributed(Activation('sigmoid'))(attention_output)
    x = Multiply()([x, attention_weights])

    # Residual Connection after Attention Mechanism
    x = Add()([x, x_residual])

    x = TimeDistributed(FANLayer(output_dim=512, p_ratio=0.25, activation='gelu', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)))(x)

    #x = Add()([x, fan_res])

    # Handle category_input with Embedding
    embedding_dim = 64
    num_categories = 6  # Replace with the actual number of categories

    embedding_layer = TimeDistributed(Embedding(input_dim=num_categories, output_dim=embedding_dim, name='embedding'))(category_input)
    category_embedded = Reshape((-1, embedding_layer.shape[-1] * embedding_layer.shape[-2]))(embedding_layer)

    # Apply Dense layer to adjust dimensionality (if needed)
    category_embedded = TimeDistributed(Dense(x.shape[-1], activation='relu'))(category_embedded)

    # Concatenate Point and Category Representations
    concatenated = Concatenate(axis=-1)([x, category_embedded])

    # Apply dense layers with regularization
    dense_layer = Dense(128, activation='relu', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4))(concatenated)
    dense_layer = BatchNormalization()(dense_layer)
    dropout = Dropout(0.5)(dense_layer)

    outputs = TimeDistributed(Dense(1, activation='sigmoid'))(dropout)

    # Create Model
    model = Model(inputs=[point_input, category_input], outputs=outputs)
    return model

model2 = create_model(point_input_shape,category_input_shape)
model3 = create_model(point_input_shape,category_input_shape)
optimizer = Adam(learning_rate=INITIAL_LEARNING_RATE, clipnorm=0.5)
loss = BinaryFocalCrossentropy(apply_class_balancing=False,
                              alpha=0.5869,
                              gamma=2.0,
                              from_logits=False,
                              label_smoothing=0.0,
                              axis=-1,
                              reduction='sum_over_batch_size',
                              name='binary_focal_crossentropy'
                              )
_ = model2((tf.zeros((1, 6, 588, 1)), tf.zeros((1, 6, 1))))  # providing data for both inputs
model2.load_weights("best_model_file_FANFAN_2.keras")
model2.compile(optimizer=optimizer, loss=loss, metrics=["accuracy", tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])
_ = model3((tf.zeros((1, 6, 588, 1)), tf.zeros((1, 6, 1))))  # providing data for both inputs
model3.load_weights("best_model_file_FANFAN_3.keras")
model3.compile(optimizer=optimizer, loss=loss, metrics=["accuracy", tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])
```
